## 生成乱序评价数据

```python
import pandas as pd

# 合并两种方式生成的结果
df_result_prompt = pd.read_csv('result/val_result.csv')
df_result_fintune = pd.read_csv('result/val_result_server.csv')

df_result_final =  pd.merge(df_result_fintune, df_result_prompt, on=['title', 'context'])
df_result_final = df_result_final.reindex(columns=['context','title',  'chatglm_zero_shot', 'chatglm_one_shot','chatglm_least_to_most','finetuned_underfit','finetuned_justfit','finetuned_overfit'])
df_result_final.to_csv('result/val_result_final.csv', index=False)
```

```python
df_result_final
```

```python
# 将顺序打乱，生成评价的Excel
df_eval = pd.DataFrame()
df_eval['新闻内容'] = df_result_final['context']
df_eval = df_eval.reindex(columns=df_eval.columns.tolist() + ['标题1', '标题2','标题3','标题4','标题5','标题6','标题7','最佳标题','最差（有问题）标题'])
for i in range(len(df_result_final)):
    title = df_result_final.iloc[i,1:].copy()
    shuffled_title = title.sample(frac=1).reset_index(drop=True)
    for j in range(1,8):
        df_eval.iloc[i,j] = shuffled_title.iloc[j-1]
        
        
```

```python
df_eval.to_excel('result/eval.xlsx', index=False)
df_eval
```

## 统计汇总评价结果

```python
df_eval_result = pd.read_excel('result/eval_result.xlsx')
df_eval_result
```

```python
def find_columns_with_target_string(df, target):
    result_columns = []
    result_series = pd.Series([])
    df = df.iloc[:,1:].copy()

    # 遍历每个单元格
    for column in df.columns:
        # 对每个单元格应用 lambda 表达式
        matches = df[column].apply(lambda x: target in str(x))
        # 如果单元格内容包含目标字符串，则将该列的列名添加到结果列表中
        if matches.any():
            return column
```

```python
import numpy as np

title_score = {
    'title': [0,0],
    'chatglm_zero_shot': [0,0],
    'chatglm_one_shot':[0,0],
    'chatglm_least_to_most':[0,0],
    'finetuned_underfit':[0,0],
    'finetuned_justfit':[0,0],
    'finetuned_overfit':[0,0]
}


for i in range(len(df_eval_result)):
    eval_best_index = df_eval_result.iloc[i,8]
    eval_worst_indexes = str(df_eval_result.iloc[i,9]).split(',')
    
    
    best_result = df_eval_result[f'标题{eval_best_index}'].iloc[i]
    
    best_column = find_columns_with_target_string(df_result_final, best_result)
    title_score[best_column][0] += 1
    
    for index in eval_worst_indexes:
        worst_result = str(df_eval_result[f'标题{index}'].iloc[i])
        if len(worst_result)>1:
            worst_column = find_columns_with_target_string(df_result_final, worst_result)
            title_score[worst_column][1] += 1
```
